```python
# This notebook was written to be run after Ben's generate_SPECTER2_embeddings.ipynb. Ideally, it should just be one
# notebook but I was having some trouble running Ben's notebook so this takes the output csv of it

# import libraries
import pandas as pd
```

```python
# load output csv
df = pd.read_csv("2022-dhq-articles-with-abstracts-and-SPECTER-recommendations.csv")
```

```python
# Create format string
url_string = "digitalhumanities.org/dhq/vol/{0}/{1}/{2}/{2}.html"
```

```python
# Split the Volume and issue to create the urls from those strings
urls = []
for index, row in df.iterrows():
    volandissue = str(row["Volume and Issue"]).split(".")
    url = url_string.format(volandissue[0], volandissue[1], (str(row["Article ID"]).zfill(6)))
    urls.append(url)
```

```python
# Add the urls to the df
df['url'] = urls
```

```python
# Small function to zfill the zeroes to the DHQ IDs so that they are compliant with the urls
def fill(id):
    return str(id).zfill(6)
    
```

```python
# Apply function to all ID related columns
df["Recommendation 1"] = df["Recommendation 1"].apply(fill)
df["Recommendation 2"] = df["Recommendation 2"].apply(fill)
df["Recommendation 3"] = df["Recommendation 3"].apply(fill)
df["Recommendation 4"] = df["Recommendation 4"].apply(fill)
df["Recommendation 5"] = df["Recommendation 5"].apply(fill)
df["Recommendation 6"] = df["Recommendation 6"].apply(fill)
df["Recommendation 7"] = df["Recommendation 7"].apply(fill)
df["Recommendation 8"] = df["Recommendation 8"].apply(fill)
df["Recommendation 9"] = df["Recommendation 9"].apply(fill)
df["Recommendation 10"] = df["Recommendation 10"].apply(fill)
df["Article ID"] = df["Article ID"].apply(fill)
```

```python
# Output and save tsv file
df.to_csv("dhq-recs-zfill.tsv", sep="\t", index=False)
```
